```python
import pandas as pd

df_EU = pd.read_csv(r"C:\Users\User\Downloads\Daten_Aufgabe_1 - Shopauswertung.csv", sep=';')
df_EU["Datum"] = pd.to_datetime(df_EU["Datum"], format="%d.%m.%Y")


df_EU["Datum"] = df_EU["Datum"].dt.strftime('%b')

umsatz_pro_monat_produkt = df_EU.groupby(['Datum', 'Produkt'])['Umsatz'].sum().unstack().reset_index()
umsatz_pro_monat_produkt["C"] = umsatz_pro_monat_produkt["A+B+C"] - umsatz_pro_monat_produkt["A"] - umsatz_pro_monat_produkt["B"]

round(umsatz_pro_monat_produkt)
```
